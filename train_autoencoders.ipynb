```python
import os
import shutil
import cv2
import numpy as np
import matplotlib.pyplot as plt
import tensorflow as tf
from tensorflow.keras import layers, models, callbacks
from tensorflow.keras.preprocessing.image import ImageDataGenerator
from tensorflow.keras.models import load_model
from sklearn.model_selection import train_test_split
from sklearn.metrics import (accuracy_score, precision_score, classification_report,
                             recall_score, f1_score, confusion_matrix,
                             ConfusionMatrixDisplay, roc_curve, auc)
from google.colab import drive

# ==========================================
# 1. SETUP & MOUNT DRIVE
# ==========================================
drive.mount('/content/drive', force_remount=True)

drive_project_path = '/content/drive/MyDrive/DL Project'
drive_eyes_path = os.path.join(drive_project_path, 'eyes_data')
drive_yawn_path = os.path.join(drive_project_path, 'yawn_data')

local_eyes_path = '/content/local_eyes_data'
local_yawn_path = '/content/local_yawn_data'

eye_model_path = os.path.join(drive_project_path, 'eye_autoencoder.h5')
yawn_model_path = os.path.join(drive_project_path, 'yawn_autoencoder.h5')

# ==========================================
# 2. DELETE OLD MODELS TO START FRESH
# ==========================================
if os.path.exists(eye_model_path):
    os.remove(eye_model_path)
    print("Deleted old Eye model!")

if os.path.exists(yawn_model_path):
    os.remove(yawn_model_path)
    print("Deleted old Yawn model!")

# ==========================================
# 3. COPY DATA TO LOCAL STORAGE
# ==========================================
print("Copying Eyes data...")
if os.path.exists(local_eyes_path): shutil.rmtree(local_eyes_path)
shutil.copytree(drive_eyes_path, local_eyes_path)

print("Copying Yawn data...")
if os.path.exists(local_yawn_path): shutil.rmtree(local_yawn_path)
shutil.copytree(drive_yawn_path, local_yawn_path)

# ==========================================
# 4. HELPER FUNCTIONS (LOADING & EVALUATION)
# ==========================================
def load_images_from_folder(folder, size=(64,64)):
    images = []
    if not os.path.exists(folder):
        print(f"Warning: Folder not found {folder}")
        return np.array([])
    files = [f for f in os.listdir(folder) if f.lower().endswith(('.jpg','.jpeg','.png'))]
    for f in files:
        img = cv2.imread(os.path.join(folder, f))
        if img is not None:
            img = cv2.resize(img, size)
            img = img.astype("float32") / 255.0
            images.append(img)
    return np.array(images)

def calculate_mse_batch(model, images):
    if len(images) == 0: return np.array([])
    reconstructions = model.predict(images, verbose=0)
    mses = np.mean(np.power(images - reconstructions, 2), axis=(1, 2, 3))
    return mses

def evaluate_arrays(model, normal_imgs, anomaly_imgs, model_name):
    print(f"\n{'='*50}\n Evaluating: {model_name}\n{'='*50}")
    normal_mses  = calculate_mse_batch(model, normal_imgs)
    anomaly_mses = calculate_mse_batch(model, anomaly_imgs)

    all_mses   = np.concatenate([normal_mses, anomaly_mses])
    all_labels = np.concatenate([np.zeros(len(normal_mses)), np.ones(len(anomaly_mses))])

    best_f1, best_threshold = 0, 0
    for t in np.linspace(all_mses.min(), all_mses.max(), 200):
        preds = (all_mses > t).astype(int)
        f1 = f1_score(all_labels, preds, zero_division=0)
        if f1 > best_f1:
            best_f1 = f1
            best_threshold = t

    preds = (all_mses > best_threshold).astype(int)
    report = classification_report(all_labels, preds, target_names=['Normal', 'Anomaly'])
    fpr, tpr, _ = roc_curve(all_labels, all_mses)
    roc_auc = auc(fpr, tpr)

    # Plotting
    fig, axes = plt.subplots(1, 3, figsize=(18, 5))
    axes[0].hist(normal_mses, bins=50, alpha=0.7, color='green', label='Normal (Unseen)')
    axes[0].hist(anomaly_mses, bins=50, alpha=0.7, color='red', label='Anomaly')
    axes[0].axvline(best_threshold, color='black', linestyle='--', label=f'Threshold: {best_threshold:.4f}')
    axes[0].set_title('MSE Distribution')
    axes[0].legend()

    disp = ConfusionMatrixDisplay(confusion_matrix=confusion_matrix(all_labels, preds), display_labels=['Normal', 'Anomaly'])
    disp.plot(ax=axes[1], colorbar=False, cmap='Blues')
    axes[1].set_title('Confusion Matrix')

    axes[2].plot(fpr, tpr, color='blue', label=f'AUC = {roc_auc:.3f}')
    axes[2].plot([0,1], [0,1], color='gray', linestyle='--')
    axes[2].set_title('ROC Curve')
    axes[2].legend()
    plt.show()

    return {'report': report, 'threshold': best_threshold, 'auc': roc_auc}

# ==========================================
# 5. AUTOENCODER ARCHITECTURE & AUGMENTATION
# ==========================================
def build_autoencoder():
    input_img = layers.Input(shape=(64, 64, 3))
    x = layers.Conv2D(32, (3, 3), activation='relu', padding='same')(input_img)
    x = layers.BatchNormalization()(x)
    x = layers.MaxPooling2D((2, 2), padding='same')(x)
    x = layers.Dropout(0.2)(x)
    x = layers.Conv2D(64, (3, 3), activation='relu', padding='same')(x)
    x = layers.BatchNormalization()(x)
    x = layers.MaxPooling2D((2, 2), padding='same')(x)
    x = layers.Dropout(0.2)(x)
    x = layers.Conv2D(128, (3, 3), activation='relu', padding='same')(x)
    encoded = layers.MaxPooling2D((2, 2), padding='same')(x)

    x = layers.Conv2D(128, (3, 3), activation='relu', padding='same')(encoded)
    x = layers.UpSampling2D((2, 2))(x)
    x = layers.Conv2D(64, (3, 3), activation='relu', padding='same')(x)
    x = layers.UpSampling2D((2, 2))(x)
    x = layers.Conv2D(32, (3, 3), activation='relu', padding='same')(x)
    x = layers.UpSampling2D((2, 2))(x)
    decoded = layers.Conv2D(3, (3, 3), activation='sigmoid', padding='same')(x)

    autoencoder = models.Model(input_img, decoded)
    autoencoder.compile(optimizer='adam', loss='mse')
    return autoencoder

train_datagen = ImageDataGenerator(
    rotation_range=15, width_shift_range=0.1, height_shift_range=0.1,
    brightness_range=[0.5, 1.5], horizontal_flip=True, zoom_range=0.15,
    shear_range=0.1, fill_mode='nearest'
)

# ==========================================
# 6. TRAIN & EVALUATE EYE AUTOENCODER
# ==========================================
print("\n--- Processing Eyes Data ---")
eye_open_path = os.path.join(local_eyes_path, 'open_eyes')
eye_closed_path = os.path.join(local_eyes_path, 'closed_eyes')

X_eyes_normal = load_images_from_folder(eye_open_path)
X_eyes_anomaly = load_images_from_folder(eye_closed_path)

# STRICT 80/20 SPLIT (random_state=42 ensures reproducibility)
X_eye_train, X_eye_test_norm = train_test_split(X_eyes_normal, test_size=0.2, random_state=42)

print(f"Eye Training Samples (Normal): {len(X_eye_train)}")
print(f"Eye Testing Samples (Normal Unseen): {len(X_eye_test_norm)}")
print(f"Eye Testing Samples (Anomaly Unseen): {len(X_eyes_anomaly)}")

eye_ae = build_autoencoder()
eye_callbacks = [
    callbacks.EarlyStopping(monitor='val_loss', patience=7, restore_best_weights=True, verbose=1),
    callbacks.ModelCheckpoint(eye_model_path, monitor='val_loss', save_best_only=True, verbose=1),
    callbacks.ReduceLROnPlateau(monitor='val_loss', factor=0.5, patience=3, min_lr=1e-6, verbose=1)
]

print("\n--- Training Eye Autoencoder ---")
eye_ae.fit(
    train_datagen.flow(X_eye_train, X_eye_train, batch_size=32),
    validation_data=(X_eye_test_norm, X_eye_test_norm),
    epochs=80, callbacks=eye_callbacks
)

eye_results = evaluate_arrays(eye_ae, X_eye_test_norm, X_eyes_anomaly, 'Eye Autoencoder')

# ==========================================
# 7. TRAIN & EVALUATE YAWN AUTOENCODER
# ==========================================
print("\n--- Processing Yawn Data ---")
yawn_not_path = os.path.join(local_yawn_path, 'not_yawn')
yawn_yes_path = os.path.join(local_yawn_path, 'yawn')

X_yawn_normal = load_images_from_folder(yawn_not_path)
X_yawn_anomaly = load_images_from_folder(yawn_yes_path)

# STRICT 80/20 SPLIT
X_yawn_train, X_yawn_test_norm = train_test_split(X_yawn_normal, test_size=0.2, random_state=42)

print(f"Yawn Training Samples (Normal): {len(X_yawn_train)}")
print(f"Yawn Testing Samples (Normal Unseen): {len(X_yawn_test_norm)}")
print(f"Yawn Testing Samples (Anomaly Unseen): {len(X_yawn_anomaly)}")

yawn_ae = build_autoencoder()
yawn_callbacks = [
    callbacks.EarlyStopping(monitor='val_loss', patience=7, restore_best_weights=True, verbose=1),
    callbacks.ModelCheckpoint(yawn_model_path, monitor='val_loss', save_best_only=True, verbose=1),
    callbacks.ReduceLROnPlateau(monitor='val_loss', factor=0.5, patience=3, min_lr=1e-6, verbose=1)
]

print("\n--- Training Yawn Autoencoder ---")
yawn_ae.fit(
    train_datagen.flow(X_yawn_train, X_yawn_train, batch_size=32),
    validation_data=(X_yawn_test_norm, X_yawn_test_norm),
    epochs=80, callbacks=yawn_callbacks
)

yawn_results = evaluate_arrays(yawn_ae, X_yawn_test_norm, X_yawn_anomaly, 'Yawn Autoencoder')

# ==========================================
# 8. FINAL SUMMARY
# ==========================================
print("\n" + "="*50)
print(" FINAL SUMMARY ".center(50, "="))
print("="*50)

print(f"\n[EYE AUTOENCODER]")
print(f"Best Threshold: {eye_results['threshold']:.6f}")
print(eye_results['report'])

print(f"\n[YAWN AUTOENCODER]")
print(f"Best Threshold: {yawn_results['threshold']:.6f}")
print(yawn_results['report'])

print("\n" + "="*50)
print(f"Models successfully saved to your Drive!")
print(f"Update your main_system.py thresholds with:")
print(f"MSE_EYE_THRESH  = {eye_results['threshold']:.6f}")
print(f"MSE_YAWN_THRESH = {yawn_results['threshold']:.6f}")
```
